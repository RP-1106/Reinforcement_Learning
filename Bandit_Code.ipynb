```python
import numpy as np
import pandas as pd
from math import *
```

```python
df = pd.read_csv("imdb_top_1000.csv")
df.head()
```

```python
df = df.drop(["Poster_Link","Released_Year","Certificate","Overview",
              "Director","Star1","Star2","Star3","Star4","No_of_Votes",
             "Gross"],axis=1)
```

```python
df
```

```python
# Number of arms = no.of movies
k = len(df)
```

```python
epsilon = 0.1  # Exploration rate
num_trials = 1000  # Number of user interactions
```

```python
# Initialize rewards and counts
rewards = df['IMDB_Rating'].values  
total_counts = np.zeros(k)  # Number of times each arm has been selected
estimated_values = np.zeros(k)  # Estimated values for each arm
```

```python
total_reward = 0
```

```python
for _ in range(num_trials):
    if np.random.rand() < epsilon:
        selected_arm = np.random.randint(k)
    else:
        selected_arm = np.argmax(estimated_values)

    reward = rewards[selected_arm]

    total_counts[selected_arm] += 1
    estimated_values[selected_arm] += (reward - estimated_values[selected_arm]) / total_counts[selected_arm]

    total_reward += reward
```

```python
# Get the top recommended titles based on estimated values
top_recommendations = df.loc[np.argsort(estimated_values)[::-1]]

print("Total Reward :", round(total_reward,2))
print("\nTop Recommendations:")
print(top_recommendations[['Series_Title', 'IMDB_Rating', 'Genre']].head(10)) 
```
